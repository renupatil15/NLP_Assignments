```python
import nltk
from nltk.wsd import lesk
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize

print("========================================")
print(" Word Sense Disambiguation Application")
print(" Type 'exit' to stop")
print("========================================")

while True:

    sentence = input("\nEnter a sentence: ")

    if sentence.lower() == "exit":
        break

    word = input("Enter the ambiguous word from the sentence: ")

    if word.lower() == "exit":
        break

    # Tokenize sentence
    tokens = word_tokenize(sentence)

    # Apply Lesk algorithm
    sense = lesk(tokens, word)

    if sense:
        print("\nMost Appropriate Sense:", sense.name())
        print("Definition:", sense.definition())

        examples = sense.examples()
        if examples:
            print("Example Usage:", examples)
    else:
        print("\nNo clear sense found for the given word.")

print("\nProgram Ended Successfully")
```
